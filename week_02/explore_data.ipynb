```python
import pandas as pd
import numpy as np
```

```python
country_codes = pd.read_json("data/country_codes.json")
new= {"Code":"XK", "Name":"Kosovo"}
country_codes = country_codes.append(new,ignore_index=True)
country_codes
```

```python
revenue = pd.read_csv("data/all_revenue.csv")
revenue = revenue[revenue["Country"].notnull()]
revenue = revenue.sort_index()
revenue = revenue.merge(country_codes, how = "left", left_on="Country", right_on="Code")
revenue = revenue.set_index(["Name", "categories"])
revenue = revenue.drop("Code", axis=1)
revenue
```

```python
np.unique(np.array([i for i,_ in revenue.index]))
```

```python
expense = pd.read_csv("data/all_expense.csv")
expense = expense.sort_index()
expense = expense.merge(country_codes, how = "left", left_on="Country", right_on="Code")
expense = expense.set_index(["Name", "categories"])
expense = expense.drop("Code", axis=1)
expense
```

```python
np.unique(np.array([i for i,_ in expense.index]))
```

```python
revenue_countries = set(np.unique(np.array([i for i,_ in expense.index])))
expense_countries = set(np.unique(np.array([i for i,_ in expense.index])))
```

```python
for country in expense_countries:
    if expense.loc[country].iloc[1:, 1:].isnull().values.all():
        expense = expense.drop(index=country)
```

```python
expense_countries = set(np.unique(np.array([i for i,_ in expense.index])))
dropping = revenue_countries - expense_countries
```

```python
len(expense_countries)
```

```python
for country in revenue_countries:
    if country in dropping:
        revenue = revenue.drop(index = country)
```

```python
revenue = revenue.sort_index()
expense = expense.sort_index()
```

```python
revenue.to_csv("data/cleaned_revenue.csv")
expense.to_csv("data/cleaned_expense.csv")
```
